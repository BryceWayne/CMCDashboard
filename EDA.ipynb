```python
# dataframe.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle, os

cwd = os.getcwd()
with open(f'{cwd}/data/test.pkl', 'rb') as f:
    data = pickle.load(f)

history = data['History']

df = pd.DataFrame(history)
df.index = df['timestamp']
df.plot(x='timestamp', y='BTC', figsize=(10,6), rot=90)
plt.show()
```

```python
ticker = 'XVG'
coin = df[[f'{ticker}']]
coin['SMA_30'] = coin['XVG'].rolling(window=30).mean()
coin['EMA_30'] = coin['XVG'].ewm(span=30).mean()
coin
```
